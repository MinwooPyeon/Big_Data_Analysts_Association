```python
import pandas as pd
import numpy as np

from konlpy.tag import Okt
k = Okt()
```

```python
df = pd.read_csv('movie_rv.csv')
df
```

```python
sample = df[0:7500]
df1 = sample['document']
```

```python
df1
```

```python
# 모든 리뷰를 모아 하나의 문장으로 만들기

combined_text = ' '.join(df1)    
```

```python
print(combined_text)
```

```python
k = Okt()
```

```python
n = k.nouns(combined_text)
print(n);
```

```python
pos = k.pos(combined_text)
v = [word for word, pos in pos if pos.startswith('Verb')]
print(v)
```

```python
from collections import Counter
```

```python
n_result = Counter(n)
sorted_result = sorted(n_result.items(), key=lambda x: x[1], reverse=True)
sorted_result # good : 최고, bad : 쓰레기
```

```python
v_result = Counter(v)
sorted_result = sorted(v_result.items(), key=lambda x: x[1], reverse=True)
sorted_result
```

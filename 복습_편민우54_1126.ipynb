```python
import matplotlib.gridspec as gridspec
grid = gridspec.GridSpec(2,2) # 4ê°€ì§€ ê·¸ëž˜í”„
plt.figure(figsize=(16,10))
plt.subplots_adjust(wspace= 0.5, hspace=0.4) # ê·¸ëž˜í”„ê°€ ì—¬ëŸ¬ ê°œ ê·¸ë ¤ì§ˆ ê²½ìš°ì˜ ì¶©ëŒ ë°©ì§€

mpg_features = ['manufacturer','model','category','drv'] # ë²”ì£¼í˜• ë°ì´í„°ë§Œ ê°€ì§€ê³  ê·¸ëž˜í”„ ê·¸ë¦¬ê¸°
# sns countplot ê·¸ë¦¬ê¸°!

for idx, feature in enumerate(mpg_features):
    ax = plt.subplot(grid[idx])
    sns.countplot(x = feature,
                 data= df,
                 hue = 'drv',
                 palette = 'pastel',
                 ax =ax)
    plt.xticks(rotation=90) #xì¶•ì˜ ê°’ ì¶•ì „í™˜
    ax.set_title(f'{feature} Distribution')
```

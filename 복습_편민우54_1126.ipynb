```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib.gridspec as gridspec
grid = gridspec.GridSpec(2,2) # 4가지 그래프
plt.figure(figsize=(16,10))
plt.subplots_adjust(wspace= 0.5, hspace=0.4) # 그래프가 여러 개 그려질 경우의 충돌 방지

mpg_features = ['manufacturer','model','category','drv'] # 범주형 데이터만 가지고 그래프 그리기
# sns countplot 그리기!

for idx, feature in enumerate(mpg_features):
    ax = plt.subplot(grid[idx])
    sns.countplot(x = feature,
                 data= df,
                 hue = 'drv',
                 palette = 'pastel',
                 ax =ax)
    plt.xticks(rotation=90) #x축의 값 축전환
    ax.set_title(f'{feature} Distribution')
```

#### 시계열 데이터
- 시간의 흐름을 보여주는 인덱스, datetime 이용해 시계열 데이터를 확인할 수 있음
- 기존 분석과는 다름
- 보간법에서 본 것처럼, 시간의 흐름에 따라 데이터가 특성이 생성되는 것 이 시간의 흐름을 마음대로 바꾸게 되면 -> 시간의 특성이 사라지게 됨
- 예측을 진행할 때 문제가 될 수 있음

- ML(Machine Learning)에서 train_test_split
- 시계열은 랜덤하게 나눌 수가 없음
- 시계열 데이터는 time과 관련한 나눠서 데이터를 분할함

#### 시계열 데이터는
- pandas 에서 제공하는 datetime, TimeStamp, Timedelta 등 여러가지 제공
- datetime 시계열 데이터로 변환하여 많이 사용하고 있음
- 시계열 데이터를 다루는 것들 볼 예정

- datetime 부분 확인 예정
- date + time 개념으로 
- date 연, 월, 일 이루어지고, time 시간, 분, 초 , 마이크로초 등으로 이루어짐
- Timestamp : 날짜와 시간을 모두 가지고 있고 단일객체
- Timedelta : 객체를 가지고 있으면서 시간의 계산이 가능하다 ( 덧셈, 뺄셈 가능 )

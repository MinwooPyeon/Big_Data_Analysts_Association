### 데이터 시각화
- 시각화 모듈의 문법과 로직

- matplotlib, seaborn
- matplot이 불편한 부분이 있음 -> seaborn 사용
- 툴은 기본적으로 시각화 하는데 가장 많이 사용하는 툴
- 태블로(tableau)를 많이 보고 빠르게 확인

#### 시각화 하는 이유
- 시각화 기능
- 데이터 분석의 나의 주장이나, 가설을 설득하기 위한 과정
- 나의 주장과 가설 등을 가장 잘 뒷받침하고 가장 잘 보여줄 수 있음
- 변수간의 관계 확인용도 있지만, 나의 주장이 베이스로 진행됨

#### 시각화 유의사항
- 시각화는 상대방을 쉽게 설득시켜야 함
- 데이터 시각화로 잘못되게 표현하면 안됨
- 다양한 시각화 그래프가 많지만, 가장 단순하고 가장 확실하게 이해할 수 있도록 해야함
- 수치에 대한 주석처리와 숫자에 대한 정확한 표시는 기본
- 기본적인 수치와 단위에 대한 기준을 명시해야함

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

#### matplot
- Figure (큰 도화지의 개념
- ax (축의 개념)
- 도화지를 그릴 때 큰 그림을 그리고 -> 그 안에 그림을 채우는 형식 ( 시각화 모듈의 로직)
- 결론적으로 큰 그림을 그리고 안의 세부적인 부분 추가

#### 기본 문법
- plt. 이어짐

#### %matplotlib inline

#### 산점도를 가지고 그림 그리기
- 시각화 할때 데이터의 타입이 가장 중요함
- 연속형인지 범주형인지에 따라 그릴 수 있는 시각화가 달라짐

```python
plt.scatter([10,20,30,40],[50,60,70,80])
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

#### 축의 범위 설정
- x축과 y축이 내가 원하는 범위의 값만 보이도록 함
- 축의 범위를 조정 가능

- xlim, ylim

```python
plt.scatter([10,20,30,40],[50,60,70,80])
plt.xlim(0,30)
plt.ylim(50,70)
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
plt.scatter([10,20,30,40],[50,60,70,80])
plt.xlim(left=100) # 축의 시작값 설정 가능
plt.ylim(top=100) # 축의 끝값 설정 가능
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

#### 눈금값 설정
- xticks
- yticks
- 눈금값 설정 가능

```python
plt.scatter([10,20,30,40],[50,60,70,80])
plt.xticks([10,20,30,40],['A','B','C','D'])
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

#### 범례 설정
- legend 함수 사용

```python
plt.scatter([10,20,30,40],[50,60,70,80], label='x=40')
plt.xticks([10,20,30,40],['A','B','C','D'])
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

#### ptl 문법을 겹치게 사용하면 추가 확장됨

```python
plt.scatter([10,20,30,40],[50,60,70,80], label='x=40')
plt.scatter([10,20,30,40],[80,70,60,50], label='x=30')
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

- 범례의 위치 변경 가능
- left
- center
- right
- upper left
- low left
- 나머지도 가능

```python
plt.scatter([10,20,30,40],[50,60,70,80], label='x=40')
plt.scatter([10,20,30,40],[80,70,60,50], label='x=30')
plt.title('scatter')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(loc='center')
```

#### 두 가지 이상의 그래프 그리기

- 선그래프 그리기 (plot)
- plt.plot() 안에 들어가는 parameter는 조금씩 다를 수 있음

```python
plt.figure(figsize=(12,8))
plt.plot([10,20,30],[40,50,60],label='x1')
plt.plot([10,20,30],[60,50,40],label='x2')
plt.legend()
```

#### 색상, 선 스타일, 두께 설정 가능

```python
plt.figure(figsize=(12,8))
plt.plot([10,20,30],[40,50,60], color='green',label='x1')
plt.plot([10,20,30],[60,50,40],color='black',label='x2')
plt.legend()
```

```python
plt.figure(figsize=(12,8))
plt.plot([10,20,30],[40,50,60], color='green',linewidth=8, linestyle='dashed',label='x1')
plt.plot([10,20,30],[60,50,40],color='black',linewidth=4, linestyle='dotted',label='x2')
plt.legend()
```

#### 마커 설정
- 해당 값에 대한 지정 marker
- . 점, o 원, s 사각형, * 별, D 다이아몬드 등등
- marker, markersize 지정 가능

```python
plt.figure(figsize=(12,8))
plt.plot([10,20,30],[40,50,60],marker='D',markersize=10,color='green',label='x1')
plt.plot([10,20,30],[60,50,40],color='black',label='x2')
plt.legend()
```

#### 히트맵(heatmap)
- 2차원 배열로 각 요소들 색상으로 시각화하는 방법

```python
import pandas as pd
df = pd.DataFrame({'A':[4,3,2],
                  'B':[7,8,5]},
                  index = ['X','Y','Z'])
```

```python
sns.heatmap(df)
```

- 데이터 표시하고 싶은 경우
- annot 인자 True
- fmt 사용하면 'd', 정수, 실수 등 표현 가능
- cmap 색상 조합을 다양하게 표현

```python
sns.heatmap(df, annot = True, fmt='.1f')
```

```python
sns.heatmap(df, annot = True, fmt='.1f',linewidth=2)
```

```python
sns.heatmap(df, annot = True, fmt='.1f',linewidth=2, linecolor='green')
```

```python
sns.heatmap(df, annot = True, fmt='.1f',cmap='Greys',linewidth=2, linecolor='green')
```

#### 그래프를 실제로 그려 축을 지정하여 그래프 그리기

```python
x1 = np.arange(1,10)
x2 = np.arange(1,30)
y1 = x1*2
y2 = x2*3
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,8))

ax[0].plot(x1,y1,color='blue')
ax[1].bar(x2,y2,color='green')

ax[0].set_xlabel('ax[0] axis')
ax[1].set_xlabel('ax[1] axis')

ax[0].legend()
ax[1].legend()
```

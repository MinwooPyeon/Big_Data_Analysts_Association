### Groupby

```python
# 바로 컬럼을 잡고 통계치를 확인 가능
df.groupby('year')
```

```python
df.groupby('year').lifeExp.sum()
```

```python
# 평균 확인가능
df.groupby('year').lifeExp.mean()
```

```python
# 2가지 이상을 잡을 경우
df.groupby('year')[['lifeExp','pop']].mean()
```

```python
# df.groupby('year').lifeExp.agg
# -> 원하는 함수를 넣어서 사용 가능
```

```python
# 원하는 함수 만들어서 넣기
def my_mean(values):
    n = len(values)
    sum_1 = 0
    for value in values:
        sum_1 += value
    return sum_1/n
```

```python
df.groupby('year').lifeExp.agg(my_mean)
```

- numpy를 이용해서 추가가능
- agg 함수에 추가

```python
import numpy as np
```

```python
df.groupby('year').lifeExp.agg([np.mean,np.std,np.count_nonzero])
```

### merge
- merge 데이터 데이터를 합치는 경우에 발생
- 실무에서는 DA sql 충분히 가능하지만,
- DS 데이터셋을 정리하는 경우도 있음
- 테이블을 병합시키고 정리해서 우리가 원하는 테스트셋을 만드는 경우
- pandas 많이 사용

- 데이터간의 병합을 통해 내가 원하는 새로운 테이블 만드는 개념
- merge
- concat

```python
test0 = pd.DataFrame({'class':['a','b','c','d'],'인원' :[100,200,300,400]})
test1 = pd.DataFrame({'class':['a','b','c','d'],'벌점평균' :[5,6,7,8]})
```

```python
print(test0)
print(test1)
```

- test0과 test1
- class에 따라 인원, 벌점평균을 같이 보는 테이블 만드는 경우
- 공통의 키로 class를 잡고 병합

```python
# left join, right join, inner join, outer join
# on 공통의 칼럼

pd.merge(test0, test1, how='left',on='class')
```

```python
# concat
# 덩어리 + 덩어리
# 공통의 키를 잡는게 아닌 테이블 테이블, 덩어리 덩어리 두개가 붙는 개념

pd.concat([test0,test1],axis=0)
# 위아래로 붙인 개념, 공통적인 컬럼을 지정하지 않음
```

```python
pd.concat([test0,test1],axis=1) # 공통적인 컬럼을 지정하지 않지만 값을 알아서 찾아감
```

### 데이터를 접근 하는 방법 중 loc, iloc

```python
df_sp.loc[1] # 인덱스를 기준으로 하기 때문에 1의 인덱스 값을 찾을 수 없음
```

```python
df_sp.loc[0,'year'] # loc의 경우 컬럼명을 지정할 수 있음
```

- iloc : 행의 순서에 따라 데이터를 접근하는 방식, 컬럼의 이름을 숫자로 대체해야함

```python
df_sp.iloc[1] # 행의 순서에 따라 데이터를 접근데 인덱스 2를 가져옴
```

```python
df_sp.iloc[0,'year'] # iloc은 컬럼명 지정 불가능
```

```python
df_sp.iloc[0,-1] # 컬럼명을 숫자로 대체
```

### 컬럼의 여러가지 의미가 있는 데이터의 경우 좀 더 깔끔하게 만드는 법
- 예를 들어 통계청 데이터를 보면 컬럼들이 날짜로 나열되어 있고 이 부분을 행으로 바꾸면 더욱 깔끔하게 데이터를 볼 수 있다.

```python
pew= pd.read_csv('pew.csv')
```

```python
pew.head()
```

- melt함수
- 깔끔하게 데이터를 가공하는 문법

- 기준을 잡는 설정
- id_vars 위치를 그대로 유지한 열의 이름 지정
- value_vars 행으로 위치를 변경할 열의 이름
- var_name value_vars 위치로 변경한 열의 이름 지정
- value_name var_name 위치로 변경한 열의 이름 지정

```python
# pd,melt(데이터,id_vars='열 컬럼 지정')
pd.melt(pew,id_vars='religion')
# 열이 행으로 전환됨
```

```python
pd.melt(pew,id_vars='religion',var_name='income',value_name='count')
# 안에 있는 컬럼 값을 넣은 형태
```

```python
np.transpose(pew).head() # 그대로 행과 열만 반전하는 형태
```

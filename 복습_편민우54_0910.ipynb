정렬
1. 내림차순, 오름차순
sort_values()

```python
df.sort_values('english')
# 내림차순이 디폴트
# 오름차순인지 내림차순 인지는 항상 확인해봐야 알 수 있음
```

```python
df.sort_values('english',ascending=False)
# 뒤에 ascending=False를 사용하면 내림차순 정렬 가능
```

```python
df.sort_values('english',ascending=True)
# 뒤에 ascending=True를 사용하면 오름차순 정렬 가능
```

```python
df.sort_values('english',ascending=True)[['id','science']].query('science>=60')
# 영어를 오름차순으로 정렬하고 그 데이터로 id와 science의 열을 출력하고 science가 60점 이상인 경우만 출력
```

오름차순, 내림차순 기준이 한 개인 경우가 아님
두 개 이상의 조건으로 차순 정리하는법

```python
# df.sort_values('math')에서 하나 더 추가하면 2차원의 개념이라 리스트로 넣어야함
df.sort_values(['nclass','science'])
# nclass가 가장 앞에 있어서 메인, 서브가 science
```

```python
#오름차순 디폴트에서 내림차순도 하는법
df.sort_values(['english','nclass'],ascending=[True,False])
# english에 True가 들어가 내림차순을 하고 서브로 nclass에 False를 넣어 오름차순을 함
```

파생변수

```python
df['3_mean']=(df['math']+df['english']+df['science'])/3
```

```python
#assign 함수를 이용해 파생변수 만들기
df.assign(mean_s = (df['math']+df['english']+df['science'])/3)
```

```python
df
```

assign 함수를 이용한 여러가지 파생 변수

```python
df['3_mean'].mean()
# mean()을 사용해 평균을 구할 수 있음
# 개인 평균점수의 평균
```

```python
df.assign(mean_t = (df['math']+df['english']+df['science'])/3, inplace=False, no=1, pw=123).head(6)]
# 추가로 파생변수를 여러개 만들 수 있음, head()를 사용해 행의 개수를 정할수 있음
```

```python
import numpy as np
```

```python
# 평균을 기준으로 평균이상, 평균이하 구하기
df.assign(average=np.where(df['3_mean']>60,'평균이상','평균이하'))
```

```python
np.where(df['3_mean']>60)
# 넘파이 함수를 이용해 index로 위치를 잡고 그 잡은 위치에 값을 넣음
# 해당하는 열의 번호를 출력함
```

##### groupby
- 그룹으로 묶음
- dr.groupby(기준).agg(변수,요약통계치)결과를 확인함

```python
df.groupby('nclass')
```

```python
df
```

```python
df.groupby('nclass').agg(ncl_math=('math','mean'))
#nclass별 수학 평균
```

```python
df.groupby('nclass').agg(ncl_math=('math','sum'))
```

```python
df.groupby('nclass').agg(ncl_math = ('math','std'),
                    ncl_eng = ('english','sum'),
                    ncl_math_mean = ('math','mean'))
```

```python
df
```

```python
df_gp=df.groupby('nclass').agg(ncl_math = ('math','std'),
                        ncl_eng = ('english','sum'),
                        ncl_math_mean = ('math','mean'))
```

```python
df_gp=df.groupby('nclass').agg(ncl_math = ('math','std'),
                        ncl_eng = ('english','max'),
                        ncl_math_mean = ('math','std'))
```

```python
df_gp
```

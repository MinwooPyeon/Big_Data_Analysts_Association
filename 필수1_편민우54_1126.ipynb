```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df.info()
```

```python
import matplotlib.gridspec as gridspec
grid = gridspec.GridSpec(2,2) 
plt.figure(figsize=(16,10))
plt.subplots_adjust(wspace= 0.5, hspace=0.4) 

graph_a = ['survived','age','sex','class']

for idx, feature in enumerate(graph_a):
    ax = plt.subplot(grid[idx])

    sns.countplot(x = feature,
                 data= df,
                 hue = 'embarked',
                 palette = 'pastel',
                 ax =ax)
    plt.xticks(rotation=90) 
    ax.set_title(f'{feature} Distribution')
```

```python
graph_b = ['survived','age','sex','class']

fig, axs = plt.subplots(nrows =1, ncols = len(graph_b), figsize=(15,4))

for idx, columns in enumerate(graph_b):
    print('idx',idx)
    sns.countplot(x=columns, data=df, ax=axs[idx])
```

```python
for column in df.columns:
    if df[column].dtype == 'O':
        graph_c = df[column].value_counts()
        
        # 원형 그래프
        plt.figure(figsize=(8, 8))
        plt.pie(graph_c, labels=graph_c.index, autopct='%1.1f%%', startangle=90)
        plt.title(f'{column} graph')
        plt.show()
```
